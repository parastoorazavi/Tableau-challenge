```python
import pandas as pd
import numpy as np
import geopy.distance
```

```python
# assign the whole csv files to a variable
citibike_1 = ('Resources/JC-202001-citibike-tripdata.csv')
citibike_2 = ('Resources/JC-202002-citibike-tripdata.csv')
citibike_3 = ('Resources/JC-202003-citibike-tripdata.csv')
citibike_4 = ('Resources/JC-202004-citibike-tripdata.csv')
citibike_5 = ('Resources/JC-202005-citibike-tripdata.csv')
citibike_6 = ('Resources/JC-202006-citibike-tripdata.csv')
citibike_7 = ('Resources/JC-202007-citibike-tripdata.csv')
citibike_8 = ('Resources/JC-202008-citibike-tripdata.csv')
citibike_9 = ('Resources/JC-202009-citibike-tripdata.csv')
citibike_10 = ('Resources/JC-202010-citibike-tripdata.csv')
citibike_11 = ('Resources/JC-202011-citibike-tripdata.csv')
citibike_12 = ('Resources/JC-202012-citibike-tripdata.csv')
```

```python
# read the csv files
citibike_1_df = pd.read_csv(citibike_1)
citibike_2_df = pd.read_csv(citibike_2)
citibike_3_df = pd.read_csv(citibike_3)
citibike_4_df = pd.read_csv(citibike_4)
citibike_5_df = pd.read_csv(citibike_5)
citibike_6_df = pd.read_csv(citibike_6)
citibike_7_df = pd.read_csv(citibike_7)
citibike_8_df = pd.read_csv(citibike_8)
citibike_9_df = pd.read_csv(citibike_9)
citibike_10_df = pd.read_csv(citibike_10)
citibike_11_df = pd.read_csv(citibike_11)
citibike_12_df = pd.read_csv(citibike_12)
```

```python
citibike2020 = pd.concat([citibike_1_df, citibike_2_df,citibike_3_df,citibike_4_df,citibike_5_df,citibike_6_df, citibike_7_df,citibike_8_df,citibike_9_df,citibike_10_df,citibike_11_df,citibike_12_df], ignore_index=True)
```

```python
citibike2020
```

```python
citibike2020['gender'] = citibike2020['gender'].replace([0,1,2],["unknown","Male","Female"])
citibike2020
```

```python
# check for null values
print(citibike2020.isna().sum())
```

```python
def get_distance(row):
    coords_1 = (row["start station latitude"], row["start station longitude"])
    coords_2 = (row["end station latitude"],row["end station longitude"])
    return geopy.distance.distance(coords_1, coords_2).km
```

```python
citibike2020["distance(km)"] = citibike2020.apply(get_distance, axis=1)
```

```python
# export the citibike2020 to a csv file
citibike2020.to_csv("citibike2020.csv")
```
